```python
# change the data type of the churn column from boolean to int64 and check the dataframe again
df['Churn'] = df['Churn'].astype('int64')

df
```

```python
# get the distribution of the churn by counting how many churned and how many did not
churn_distribution = df['Churn'].value_counts()

print(churn_distribution)
```

```python
# plot the count from the result above

churn_distribution = df['Churn'].value_counts()
churn_distribution.plot(kind = 'bar')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.title('Churn Distribution')
plt.show()
```

```python
# what is the proportion of churned users in the dataframe
churned_count = df['Churn'].sum()
total_count = len(df)

proportion_churned = (churned_count / total_count) * 100

print(proportion_churned)
```

```python
# How much time (on average) do churned users spend on the phone during daytime?
churned_users = df[df['Churn'] == 1]  # Filter churned users

average_time = churned_users['Total day minutes'].mean()  
# Calculate mean of Total day minutes

print("Time spent on the phone during daytime by churned users (On Average):", average_time)
```

```python
# What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?

# Assuming your DataFrame is named 'df' and the relevant columns are 'Churn', 'International Plan', and 'International Mins'
max_intl_call = df.loc[(df['Churn'] == 0) & (df['International plan'] == 'No'), 'Total intl minutes'].max()

print(max_intl_call)
```
